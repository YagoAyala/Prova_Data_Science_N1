## Dia da semana com placares mais altos

Crie um gráfico que responda em qual dia da semana os placares foram mais altos. O conceito de "placar mais alto" é a soma dos gols feitos no jogo.

Como a quantidade de jogos não é a mesma em todos os dias da semana (finais de semana tendem a ter mais jogos que em dias da semana - o que resultaria em placares mais altos nestes dias), você deverá encontrar uma forma de equalizar esta diferença na quantidade de jogos.

A resposta deverá ser um gráfico que demonstre claramente o comparativo dos dias.

````python
# Importando as bibliotecas necessárias
import pandas as pd
import plotly.express as px

# Carregando o dataset
url = 'https://drive.google.com/uc?export=download&id=1OxgoFW-l_vxuUPaiqLJXMWv8bkED1NNy'

df = pd.read_csv(url)

# Criando uma nova coluna para o total de gols em cada partida
df['total_goals'] = df['gf'] + df['ga']

# Convertendo a coluna 'date' para formato datetime
df['date'] = pd.to_datetime(df['date'])

# Extraindo o dia da semana da coluna 'date'
df['day_of_week'] = df['date'].dt.day_name()

# Definindo a ordem dos dias da semana
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dias_pt = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']

# Garantindo que 'day_of_week' seja uma variável categórica com a ordem especificada
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=days_order, ordered=True)

# Calculando a média de gols por partida em cada dia da semana
avg_goals_per_day = df.groupby('day_of_week')['total_goals'].mean().reset_index()

# Traduzindo os dias da semana para português
avg_goals_per_day['day_of_week_pt'] = avg_goals_per_day['day_of_week'].replace(dict(zip(days_order, dias_pt)))

# Criando o gráfico interativo com Plotly
fig = px.bar(
    avg_goals_per_day,
    x='day_of_week_pt',
    y='total_goals',
    title='Média de Gols por Partida em Cada Dia da Semana',
    labels={'day_of_week_pt': 'Dia da Semana', 'total_goals': 'Média de Gols por Partida'},
    color='total_goals',
    color_continuous_scale='Turbo',
    template='plotly_dark',
    text='total_goals'
)

# Personalizando o layout do gráfico
fig.update_layout(
    xaxis_title='Dia da Semana',
    yaxis_title='Média de Gols por Partida',
    title_font_size=24,
    font=dict(size=14)
)

# Ajustando o tamanho do texto nas barras
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

# Exibindo o gráfico
fig.show()
````

## Capitães

Quantos capitães diferentes cada equipe teve durante a temporada?

Responda em forma de um gráfico comparativo, apresentando os dados o mais claramente possível.

````python
# Importing necessary libraries
import pandas as pd
import plotly.express as px

# Loading the dataset
url = 'https://drive.google.com/uc?export=download&id=1OxgoFW-l_vxuUPaiqLJXMWv8bkED1NNy'

df = pd.read_csv(url)

# Counting the number of unique captains for each team
captain_count = df.groupby('team')['captain'].nunique().reset_index()

# Renaming the columns for clarity
captain_count.columns = ['team', 'unique_captains']

# Sorting the data by the number of unique captains
captain_count = captain_count.sort_values(by='unique_captains', ascending=False)

# Creating a bar chart with Plotly and adding a tooltip
fig = px.bar(
    captain_count,
    x='team',
    y='unique_captains',
    title='Number of Different Captains Per Team in the Season',
    labels={'team': 'Team', 'unique_captains': 'Number of Different Captains'},
    color='unique_captains',
    color_continuous_scale='Viridis',
    template='plotly_dark',
    text='unique_captains'
)

# Adjusting the layout of the chart
fig.update_layout(
    xaxis_title='Team',
    yaxis_title='Number of Different Captains',
    title_font_size=24,
    font=dict(size=14),
    xaxis={'categoryorder': 'total descending'}
)

# Displaying the value on top of the bars
fig.update_traces(texttemplate='%{text:.0f}', textposition='outside')

# Showing the chart
fig.show()
````

## Formação Tática

Qual a formação tática mais comum para cada equipe de acordo com o resultado obtido (Ex: para vitórias, a formação mais comum é X, para empates Y e para Z em derrotas).

**Importante** responda de forma que seja possível fazer comparativos entre equipes.

Responda com um gráfico.

````python
# Importing necessary libraries
import pandas as pd
import plotly.express as px

# Loading the dataset
url = 'https://drive.google.com/uc?export=download&id=1OxgoFW-l_vxuUPaiqLJXMWv8bkED1NNy'

df = pd.read_csv(url)

# Grouping by team and result to find the most common formation for each team and result
common_formation = df.groupby(['team', 'result'])['formation'].agg(lambda x: x.value_counts().idxmax()).reset_index()

# Pivoting the table to have teams as rows and results as columns for better comparability
common_formation_pivot = common_formation.pivot(index='team', columns='result', values='formation').reset_index()

# Melting the dataframe to have it in long format for better plotting
common_formation_melted = common_formation_pivot.melt(id_vars='team', var_name='result', value_name='formation')

# Creating a Plotly bar chart for comparison
fig = px.bar(
    common_formation_melted,
    x='team',
    y='formation',
    color='result',
    title='Most Common Formation for Each Team Based on Match Result',
    labels={'team': 'Team', 'formation': 'Most Common Formation', 'result': 'Match Result'},
    barmode='group',
    text='formation',
    template='plotly_dark'
)

# Adjusting the layout
fig.update_layout(
    xaxis_title='Team',
    yaxis_title='Most Common Formation',
    title_font_size=24,
    font=dict(size=14),
    xaxis={'categoryorder': 'total descending'}
)

# Showing the chart
fig.show()
````

## Posse de bola

Qual a média de posse de bola da equipe mandante quando ela perdeu o jogo?

Responda em forma de gráfico para poder comparar entre todas as equipes.

````python
# Importing necessary libraries
import pandas as pd
import plotly.express as px

# Loading the dataset
url = 'https://drive.google.com/uc?export=download&id=1OxgoFW-l_vxuUPaiqLJXMWv8bkED1NNy'

df = pd.read_csv(url)

# Filtering the dataset for matches where the home team lost
home_loss_df = df[(df['venue'] == 'Home') & (df['result'] == 'L')]

# Calculating the average possession for home teams that lost
avg_possession_home_loss = home_loss_df.groupby('team')['poss'].mean().reset_index()

# Renaming the columns for clarity
avg_possession_home_loss.columns = ['team', 'average_possession']

# Sorting the data for better visualization
avg_possession_home_loss = avg_possession_home_loss.sort_values(by='average_possession', ascending=False)

# Creating a bar chart for comparison
fig = px.bar(
    avg_possession_home_loss,
    x='team',
    y='average_possession',
    title='Média de Posse de Bola das Equipes Mandantes nas Derrotas',
    labels={'team': 'Equipe', 'average_possession': 'Posse de Bola Média (%)'},
    text='average_possession',
    template='plotly_dark'
)

# Adjusting the layout
fig.update_layout(
    xaxis_title='Equipe',
    yaxis_title='Posse de Bola Média (%)',
    title_font_size=24,
    font=dict(size=14),
    xaxis={'categoryorder': 'total descending'}
)

# Displaying the value on top of the bars
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

# Showing the chart
fig.show()
````

## Público

Crie uma visualização que apresente o público total de cada equipe, e também um detalhamento por jogos.

Apresentar os dados usando a técnica "DrillDown" será considerado um "plus". A visualização DrillDown permite que gráficos sejam detalhados com cliques em itens do gráfico.

Um exemplo com tabelas pode ser encontrado [aqui](https://dash.plotly.com/dash-ag-grid/enterprise-master-detail).

Um exemplo usando gráficos pode ser encontrado [aqui](https://community.plotly.com/t/show-and-tell-drill-down-functionality-in-dash-using-callback-context/54403?u=atharvakatre)

````python
!pip install dash dash-bootstrap-components
````

````python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Carregar o dataset
url = 'https://drive.google.com/uc?export=download&id=1OxgoFW-l_vxuUPaiqLJXMWv8bkED1NNy'

df = pd.read_csv(url)

# Filtrar dados para público presente em cada jogo
df_filtered = df[['team', 'attendance', 'date', 'opponent', 'venue']].dropna()

# Aplicativo Dash
app = dash.Dash(__name__)

# Layout do aplicativo
app.layout = html.Div([
    html.H1("Público Total por Equipe e Detalhamento por Jogos"),
    dcc.Graph(id='total-attendance-graph'),
    dcc.Graph(id='match-details-graph'),
])

# Função para criar o gráfico de público total por equipe
@app.callback(
    Output('total-attendance-graph', 'figure'),
    Input('total-attendance-graph', 'clickData')
)
def update_total_attendance_graph(clickData):
    team_total_attendance = df_filtered.groupby('team')['attendance'].sum().reset_index()
    fig = px.bar(team_total_attendance, x='team', y='attendance',
                 title='Público Total por Equipe',
                 labels={'team': 'Equipe', 'attendance': 'Público Total'})
    fig.update_layout(clickmode='event+select')
    return fig

# Função para detalhamento por jogos ao clicar na equipe
@app.callback(
    Output('match-details-graph', 'figure'),
    [Input('total-attendance-graph', 'clickData')]
)
def update_match_details_graph(clickData):
    if clickData is None:
        # Se nenhum time for clicado, não mostrar detalhamento
        return px.bar(title="Clique em uma equipe para ver os detalhes por jogo")

    team_clicked = clickData['points'][0]['x']
    team_data = df_filtered[df_filtered['team'] == team_clicked]

    fig = px.bar(team_data, x='date', y='attendance', color='opponent',
                 title=f'Público por Jogos da Equipe: {team_clicked}',
                 labels={'date': 'Data', 'attendance': 'Público', 'opponent': 'Oponente'})
    return fig

# Rodar o aplicativo
if __name__ == '__main__':
    app.run_server(debug=True)
````
